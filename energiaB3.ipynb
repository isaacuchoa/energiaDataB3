```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv('caseEnergiaB3.csv')
dados
```

```python
# Alterando tipo de dados e substituindo virgula por ponto

colunas = ['Petrobras','Aeris Energy','Alupar','AES Brasil','Cemig','Cesp'] 
dados[colunas] = dados[colunas].apply(lambda x: x.str.replace(',', '.').astype('float'))
```

```python
dados
```

```python
dados.info()
```

```python
# pip install -U python-dateutil
'''
 atualizaÃ§Ã£o da lib dateutil para resolver conflitos de data
'''
```

```python
# Alterando coluna de data para o formato datetime

dados['Data'] = pd.to_datetime(dados['Data']).dt.strftime('%d%m%Y')

dados
```

```python
# Setar o index para trabalhar com series temporais

dados.set_index('Data', inplace=True)
dados.head()
```

```python
# GrÃ¡fico

#Estilo do GrÃ¡fico
plt.style.use('seaborn-darkgrid')

# Tamanho
plt.figure(figsize=(16,8))

# Titulo
plt.title("AnÃ¡lise de aÃ§Ãµes de empresa de energia", loc='left', fontsize=18, fontweight=0)

# Plot da Petrobras
plt.plot(dados.index, dados['Petrobras']);

```
