```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv('caseEnergiaB3.csv')
dados
```

```python
# Alterando tipo de dados e substituindo virgula por ponto

colunas = ['Petrobras','Aeris Energy','Alupar','AES Brasil','Cemig','Cesp'] 
dados[colunas] = dados[colunas].apply(lambda x: x.str.replace(',', '.').astype('float'))
```

```python
# Criar coluna ano e mês

dados['mês'] = pd.to_datetime(dados['Data']).dt.strftime('%m')
dados['Ano'] = pd.to_datetime(dados['Data']).dt.strftime('%Y')

#Excluir coluna Data
dados = dados.drop(columns=['Data'])
dados.head()
```
